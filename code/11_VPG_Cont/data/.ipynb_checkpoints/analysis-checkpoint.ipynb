# Buyer Surplus Fraction

```python
frac_profit = np.empty((3000, 15, 3))
frac_prices = np.empty((3000, 15, 3))
frac_sales = np.empty((3000, 15, 3))
std_profit = np.empty((3000, 15, 3))
std_price = np.empty((3000, 15, 3))
spread = np.empty((3000, 15, 3))

for j, idx in enumerate([23, 15, 1]):
    roundFiles = glob.glob(f'round_single_{idx}_*')
    logFiles = glob.glob(f'log_single_{idx}_*')
    print(len(roundFiles))
    for i in range(len(roundFiles)):
        df = pd.read_csv(logFiles[i])
        df2 = pd.read_csv(roundFiles[i])
        buyerFrac = df.groupby('rnd').sum().bprofit / df2.buyerSurplus 
        priceFrac = df.groupby('rnd').mean().price / df2.peq 
        saleFrac = df.groupby('rnd').sum().sale / df2.qeq 
        frac_profit[:, i, j] = buyerFrac.rolling(window=100).mean().values
        frac_prices[:, i, j] = priceFrac.rolling(window=100).mean().values
        frac_sales[:, i, j] = saleFrac.rolling(window=100).mean().values
        std_profit[:, i, j] = df.groupby('rnd').std().bprofit.rolling(window=100).mean().values
        std_price[:, i, j] = df.groupby('rnd').std().price.rolling(window=100).mean().values
        a = df.groupby(['rnd', 'currentBidIdx']).sum().bprofit.groupby(['rnd']).max()
        b = df.groupby(['rnd', 'currentBidIdx']).sum().bprofit.groupby(['rnd']).min()
        c = a-b
        spread[:, i, j] = c.rolling(window=100).mean().values
```

```python
plt.plot(np.nanmean(frac_profit[:, 0:13, 0],1), label = 'Full Disclosure')
plt.plot(np.nanmean(frac_profit[:, :, 1],1), label = 'Winner Identity')
plt.plot(np.nanmean(frac_profit[:, :, 2],1), label = 'No Disclosure')
plt.title("Buyer Surplus Fraction: Actual to Theoretical")
plt.legend()
plt.show()

plt.plot(np.nanmean(frac_prices[:, 0:13, 0],1), label = 'Full Disclosure')
plt.plot(np.nanmean(frac_prices[:, :, 1],1), label = 'Winner Identity')
plt.plot(np.nanmean(frac_prices[:, :, 2],1), label = 'No Disclosure')
plt.title("Price Fraction: Actual to Theoretical")
plt.legend()
plt.show()


plt.plot(np.nanmean(std_profit[:, 0:13, 0],1), label = 'Full Disclosure')
plt.plot(np.nanmean(std_profit[:, :, 1],1), label = 'Winner Identity')
plt.plot(np.nanmean(std_profit[:, :, 2],1), label = 'No Disclosure')
plt.title("Avg. Volatility in Profits Per Round")
plt.legend()
plt.show()

plt.plot(np.nanmean(std_price[:, 0:13, 0],1), label = 'Full Disclosure')
plt.plot(np.nanmean(std_price[:, :, 1],1), label = 'Winner Identity')
plt.plot(np.nanmean(std_price[:, :, 2],1), label = 'No Disclosure')
plt.title("Avg. Volatility in Profits Per Round")
plt.legend()
plt.show()

plt.plot(np.nanmean(spread[:, 0:13, 0],1), label = 'Full Disclosure')
plt.plot(np.nanmean(spread[:, :, 1],1), label = 'Winner Identity')
plt.plot(np.nanmean(spread[:, :, 2],1), label = 'No Disclosure')
plt.title("Max and Min Profit Spread Per Round")
plt.legend()
plt.show()
```
